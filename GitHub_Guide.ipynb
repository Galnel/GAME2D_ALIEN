# 🧠 1️⃣ O que é o GitHub?

O GitHub é uma plataforma online para hospedar repositórios Git.

Ele usa o Git, que controla versões do seu código.

# 🚀 2️⃣ Criando um repositório no GitHub

Vá para:
https://github.com

Clique em New repository

Escolha:

- Nome

- Public ou Private

- Clique em Create repository

- Pronto ✅

# 💻 3️⃣ Conectar seu projeto local ao GitHub

Dentro da pasta do seu projeto:

1. Inicializar Git (se ainda não tiver)

```python
git init
```

2. Conectar ao repositório remoto

     -  Copie a URL do GitHub e rode:

```python
git remote add origin https://github.com/SEU_USUARIO/NOME_REPO.git
```

Verifique:

git remote -v

# 📦 4️⃣ Primeiro envio de código

```python
git add .
git commit -m "Primeiro commit"
git branch -M main
git push -u origin main
```

Agora seu código está no GitHub 🚀

# 🌿 5️⃣ Trabalhando com branches (forma correta)

Nunca trabalhe direto na main.

Criar nova branch:

```python
git switch -c feat/nova-feature
```

Trabalhe normalmente:

```python
git add .
git commit -m "Implementa nova feature"
git push -u origin feat/nova-feature
```

# 🔁 6️⃣ Atualizar seu projeto

Para baixar mudanças do GitHub:

```python
git pull
```

Ou primeiro atualizar referências:

```python
git fetch
```

# 🔀 7️⃣ Pull Request (fluxo profissional)

Depois de subir uma branch:

1. Vá no GitHub

2. Clique em Compare & Pull Request

3. Revise mudanças

4. Clique em Merge

Isso junta sua branch com a main.

### 🛠 Fluxo diário resumido

```python
git switch -c feat/coisa-nova
# trabalha
git add .
git commit -m "descrição"
git push
```

Depois faz Pull Request.

### 📌 Comandos essenciais
Ação	Comando
Ver status	git status
Ver branches	git branch
Criar branch	git switch -c nome
Trocar branch	git switch nome
Enviar código	git push
Atualizar	git pull
